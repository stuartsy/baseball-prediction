```python
# function that gives us the score of a player for one season
def score(row):
    return row['R'] + row['H'] + row['2B'] + (2 * row['3B']) + (3 * row['HR']) + row['RBI'] + row['SB']
```

```python
# calculate score
batting_df['score'] = batting_df.apply(lambda row: score(row), axis=1)
batting_df[:5]
# sort by score
# batting_df.sort('score', ascending=False)
```

```python
# merge repeated entries from stints
pitchers_grouped = pitchers.groupby(pitchers.index).agg({'playerID': np.max, 'yearID': np.max, 'teamID': np.max, 'W': np.sum, 'L': np.sum, 'G': np.sum, 'SHO': np.sum, 'SV': np.sum, 'IPouts': np.sum, 'H': np.sum, 'ER': np.sum, 'HR': np.sum, 'BB': np.sum, 'SO': np.sum, 'BAOpp': np.mean, 'R': np.sum})
pitchers_grouped[:10]
```

```python
def pitcher_ERA(row):
    if row['IPouts'] == 0:
        return 15
    return (row['ER'] / row['IPouts']) * 27
    
def pitcher_score(row):
    return row['IPouts'] - (3 * row['ER']) - row['H'] - row['BB'] + row['SO'] + (5 * row['W'])
```

```python
pitchers_grouped['ERA'] = pitchers_grouped.apply(lambda row: pitcher_ERA(row), axis=1)

pitchers_grouped['score'] = pitchers_grouped.apply(lambda row: pitcher_score(row), axis=1)

pitchers_grouped.sort('score', ascending=False, inplace=True)

pitchers_grouped[:10]
```

```python
# testing on 2010 - 2014

# we take 2010's statistics, and try to predict 2011's score
# we also get the difference between 2009 and 2010 scores as a predictor for 2011 score

prediction_year = 2010

pitchers_2009 = pitchers_complete[pitchers_complete["yearID"] == prediction_year - 2]
pitchers_2009 = pitchers_2009.rename(columns = {'score':'old_score'})
pitchers_2009_rel = pitchers_2009[["playerID", "old_score"]]

pitchers_2010 = pitchers_complete[pitchers_complete["yearID"] == prediction_year - 1]
pitchers_2010 = pd.merge(left=pitchers_2010, right=pitchers_2009_rel, left_on='playerID', right_on='playerID')
pitchers_2010["delta_score"] = pitchers_2010.apply(lambda row: row['score'] - row['old_score'], axis=1)
pitchers_2010.drop(['score', 'old_score'], axis=1, inplace=True)

pitchers_2011 = pitchers_complete[pitchers_complete["yearID"] == prediction_year]
pitchers_2011_rel = pitchers_2011[["playerID", "score"]]
pitchers_test = pd.merge(left=pitchers_2010, right=pitchers_2011_rel, left_on='playerID', right_on='playerID')
```

```python
pitchers_test["constant_factor"] = 1
pitchers_test[:10]
```

```python
batters_2009 = bats2_final[bats2_final["yearID"] == prediction_year - 2]
batters_2009 = batters_2009.rename(columns = {'score':'old_score'})
batters_2009_rel = batters_2009[["playerID", "old_score"]]

batters_2010 = bats2_final[bats2_final["yearID"] == prediction_year - 1]
batters_2010 = pd.merge(left=batters_2010, right=batters_2009_rel, left_on='playerID', right_on='playerID')
batters_2010["delta_score"] = batters_2010.apply(lambda row: row['score'] - row['old_score'], axis=1)
batters_2010.drop(['score', 'old_score'], axis=1, inplace=True)

batters_2011 = bats2_final[bats2_final["yearID"] == prediction_year]
batters_2011 = batters_2011[["playerID", "score"]]

batters_test = pd.merge(left=batters_2010, right=batters_2011, left_on='playerID', right_on='playerID')
```

```python
batters_test["constant_factor"] = 1
batters_test[:10]
```

```python
# Import the linear regression class
from sklearn.linear_model import LinearRegression
# Sklearn also has a helper that makes it easy to do cross validation
from sklearn.cross_validation import KFold
import numpy as np

# The columns we'll use to predict the target
pitcher_predictors = ["BB", "H", "IPouts", 
                      "L", "HR", "BAOpp", 
                      "SO", 'W', "SV", 
                      "R", "ER", "SHO", 
                      "weight", "height", "salary", 
                      "age", "constant_factor"]

# removed G and ERA and delta_score

# Initialize our algorithm class
alg = LinearRegression()

kf = KFold(pitchers_test.shape[0], n_folds=3, random_state=1)
pitcher_predictions = []
for train, test in kf:
    # The predictors we're using the train the algorithm.  Note how we only take the rows in the train folds.
    train_predictors = (pitchers_test[pitcher_predictors].iloc[train,:])
    # The target we're using to train the algorithm.
    train_target = pitchers_test["score"].iloc[train]
    # Training the algorithm using the predictors and target.
    alg.fit(train_predictors, train_target)
    # We can now make predictions on the test fold
    test_predictions = alg.predict(pitchers_test[pitcher_predictors].iloc[test,:])
    pitcher_predictions.append(test_predictions)
                       
pitcher_predictions = np.concatenate(pitcher_predictions, axis=0)
```

```python
prediction_diff = [abs(pitcher_predictions[i] - pitchers_test["score"][i]) for i in range(len(pitcher_predictions))]
sum(prediction_diff) / len(prediction_diff)
```
